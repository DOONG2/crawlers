```python
def getReviewResult(CODE, NAME):
    
    url1 = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code='
    url2 = '&type=after&isActualPointWriteExecute=false&isMieageSubscriptionAlready=false&isMileageSubscriptionReject=false&page='
    page = int(1)
    count = int(100)
    sym_l = []
    user_l = []
    score_l = []
    review_l = []
    reple_l = []
    not_sym_l = []
    
    while count:
        URL = url1+CODE+url2+str(page)
        t = urlopen(URL)
        html = t.read().decode("utf-8")
        soup = BeautifulSoup(html, "html5lib")
        score_result = soup.find('div', class_='score_result')
        lis = score_result.find_all('li')
        for li in lis:
            page = int(page)
            reple = li.find('div', class_='score_reple').find('p').get_text()
            score = li.find('div', class_='star_score').find('em').get_text()
            sym = li.find('div', class_='btn_area').find_all('strong')[0].get_text()
            not_sym = li.find('div', class_='btn_area').find_all('strong')[1].get_text()
            children = li.find('dt')
            child = children.findChildren()
            user = str(child[2])
            user = user[6:-7]
      
            if(reple[0:3] == "관람객"):
                reple = reple[3:]
                review = "O"
            else:
                review = "X"
            
            sym_l.append(sym)
            not_sym_l.append(not_sym)
            user_l.append(user)
            score_l.append(score)
            review_l.append(review)
            reple_l.append(reple)
        count -= 1
    
        if not count:
            break
        page += 1
        print(page)
        
    tempDic = {"sym": sym_l, "not_sym": not_sym_l, "user": user_l, "score": score_l, "review": review_l, "reple":reple_l}
    data = pd.DataFrame(tempDic)
    data.to_csv(NAME+".csv", mode='w', encoding="cp949")
  #f.close()
```
