```python

import urllib.request
import re
from bs4 import BeautifulSoup
from urllib.request import urlopen
import urllib
import sys
import pandas as pd
from pandas import DataFrame
import io
import chardet
import json
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import datetime
```

```python
def getTwitterResult(NAME, sp, ed):
    
    sp_year = int(sp[0:4])
    sp_mon = int(sp[4:6])
    sp_day = int(sp[6:8])
    ed_year = int(ed[0:4])
    ed_mon = int(ed[4:6])
    ed_day = int(ed[6:8])
    
    dt1 = datetime.datetime(sp_year, sp_mon, sp_day)
    dt2 = dt1 + datetime.timedelta(days=1)
    dt3 = datetime.datetime(ed_year, ed_mon, ed_day)
    SCROLL_PAUSE_TIME = (1)
    
    user_l = []
    text_l = []
    reple_l = []
    retweet_l = []
    date_l = []
    like_l = []
    while dt1 < dt3:
       # try:
            
            driver = webdriver.Chrome('C:/chromedriver.exe')
            driver.implicitly_wait(30)
            url1 = 'https://twitter.com/search?f=tweets&vertical=default&q="'
            url2 = '"%20since%3A'
            url3 = '%20until%3A'
            url4 = '&l=ko&src=typd&lang=ko'
            URL = url1 + NAME + url2 + dt1.strftime('%Y-%m-%d') + url3 + dt2.strftime('%Y-%m-%d') + url4
            driver.get(URL)
            last_height = driver.execute_script("return document.body.scrollHeight")
            count = 10
            while count>0:
                '''
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(SCROLL_PAUSE_TIME)
                try:
                    check = driver.find_element_by_class_name("has-more-items")
                    continue
                except:
                    break
                '''

                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(SCROLL_PAUSE_TIME)
                new_height = driver.execute_script("return document.body.scrollHeight")

                print(count)
                if new_height == last_height:
                    count -= 1
                else:
                    last_height = new_height
                    count += 1
            dt1 = dt1 + datetime.timedelta(days=1)
            dt2 = dt2 + datetime.timedelta(days=1)


            html = driver.page_source
            soup = BeautifulSoup(html, 'lxml')
            page = soup
            try:
                
                ol = soup.find("ol", class_="stream-items js-navigable-stream")
                lis = ol.find_all("div", class_="content")
            except:
                print(dt1.strftime('%Y-%m-%d') + " is null.")
                continue

            for li in lis:
                user = li.find("span", class_="username u-dir u-textTruncate").get_text()
                text = li.find("div", class_="js-tweet-text-container").get_text()
                reple = li.find("div", class_="ProfileTweet-action--reply").get_text()
                retweet = li.find("div", class_="ProfileTweet-action--retweet").get_text()
                like = li.find("div", class_="ProfileTweet-action--favorite").get_text()
                date = dt1.strftime('%Y-%m-%d')
                
                date_l.append(date)
                user_l.append(user)
                text_l.append(text)
                reple_l.append(reple)
                retweet_l.append(retweet)
                like_l.append(like)
       #    print(dt1.strftime('%Y-%m-%d') + " is null.")
          #  continue
        

    
    tempDic = { "date": date_l, "user": user_l, "text": text_l, "reple": reple_l, "retweet": retweet_l, "like": like_l, "text": text_l}

    data = pd.DataFrame(tempDic)
    data.to_csv(NAME+"("+sp+"-"+ ed+").csv", mode='w')
    
        
    
    
    
```
