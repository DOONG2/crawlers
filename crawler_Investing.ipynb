```python
#펜더멘탈 메뉴클릭
driver.find_element_by_xpath('//*[@id="filter_fundamental"]').click()
driver.implicitly_wait(30)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
table = soup.find('tbody')
```

```python
result = []
```

```python
for tr in table.find_all('tr'):
    
    name = tr.find("a").text
    url = tr.find("a").get("href")
    volume = tr.find_all("td")[2].text
    marketValue = tr.find_all("td")[3].text
    sales = tr.find_all("td")[4].text
    rateWeekProfit = tr.find_all("td")[5].text
    beta = tr.find_all("td")[6].text
    
    tempDic = { "name": name, "url":url, "volume":volume, "marketValue":marketValue, "sales":sales, "rateWeekProfit":rateWeekProfit, "beta":beta }
    result.append(tempDic)
```
